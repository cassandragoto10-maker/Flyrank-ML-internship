## Research Question

Can observable content and search-performance signals be used to prioritize pages for refresh review more effectively than simple rule-based approaches?

### Unit of Analysis

Individual content pages.

### Decision

Which pages should be reviewed first by a content or SEO team?

### Action

A reviewer examines the highest-ranked pages and decides whether to refresh, expand, optimize, monitor, protect, or leave the page unchanged.

### Cost of a Wrong Recommendation

If a declining page is not identified, traffic losses may continue and opportunities for recovery may be missed.

If a stable page is incorrectly prioritized, time and resources may be spent reviewing content that does not require immediate action.

### Why Data and Machine Learning Can Help

Large content inventories contain more pages than can realistically be reviewed manually. Data and machine learning may help identify patterns associated with declining performance and support more consistent prioritization decisions.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Rows:", len(df))
print("Columns:", len(df.columns))

print("\nCorrelation between search volume and impressions:")
print(round(df["search_volume"].corr(df["impressions_90d"]), 3))

print("\nMedian word count by trend direction:")
print(df.groupby("trend_direction")["word_count"].median())

print("\nAverage CTR by position tier:")
print(df.groupby("position_tier")["ctr"].mean().sort_values(ascending=False))
```

### Looking at the data

The starter dataset contains approximately 30,000 content pages and 44 variables describing content characteristics and performance metrics.

A preliminary analysis showed that the correlation between search volume and impressions was approximately 0.001. This suggests that keyword search volume alone is not a reliable indicator of the traffic a page actually receives.

The starter pipeline also showed that a learned model achieved a Precision@50 of approximately 0.74 compared with approximately 0.24 for a simple hand-written rule. This indicates that combining multiple signals may improve page prioritization.

The dataset also contains multiple trend categories, including pages that are increasing, stable, declining, or flat, providing useful information for prioritization analysis.
